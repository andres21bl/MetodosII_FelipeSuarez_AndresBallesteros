```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def iterativo(alfa,dt,u0,tmax):
    
    t=np.arange(0,tmax,dt)
    
    u=np.zeros_like(t)
    
    for k in range(len(u)):
        
        u[k]=(1 + alfa * dt)**k *u0
    
    return (t,u)
```

```python
def exacto(t,alfa, u0):
    return u0*np.exp(alfa*t)
```

```python
#parametros de los puntos
alpha = -1
tmax=15.
u0 = 1.0
delta_ts = [1.1, 1.5, 1.9]
t=np.linspace(0,tmax,100)
```

```python
fig, ax = plt.subplots(len(delta_ts), figsize=(8, 6), sharex=True)
for i, dt in enumerate(delta_ts):
    ax[i].plot(iterativo(alpha,dt,u0,tmax)[0],iterativo(alpha,dt,u0,tmax)[1],label=f'dt = {dt}')
    ax[i].set_ylabel('u(t)')
    ax[i].plot(t,exacto(t,alpha,u0),label="Sol.exacta")
    ax[i].legend()
plt.xlabel('Tiempo (s)')
plt.suptitle('Soluciones Oscilatorias para α = -1 y distinto Δt')
plt.show()
```
