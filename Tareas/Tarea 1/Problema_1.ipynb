```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return x**2
```

```python
def f2(x):
    return np.sin(x)
```

```python
def Der(f,x,h):
    num = -f(x+2*h)+4*f(x+h)-3*f(x)
    den = 2*h
    return num/den
```

```python
def SegDer(f,x,h):
    num = f(x+h)-2*f(x)+f(x-h)
    den = h**2
    return num/den
```

```python
x = np.linspace(0,4,100)
h = x[1]-x[0]
F1 = f1(x)
DF1 = Der(f1,x,h)
D2F1 = SegDer(f1,x,h)
plt.plot(x,F1, label="Funci√≥n")
plt.plot(x,DF1, label="Derivada")
plt.plot(x,D2F1, label="2da Derivada")
plt.grid()
plt.legend()
```

```python
x2 = np.linspace(0,2*np.pi,100)
h2 = x2[1]-x2[0]
F2 = f2(x2)
DF2 = Der(f2,x2,h2)
D2F2 = SegDer(f2,x2,h2)
plt.plot(x2,F2, label="Funci√≥n")
plt.plot(x2,DF2, label="Derivada")
plt.plot(x2,D2F2, label="2da Derivada")
plt.grid()
plt.legend()
```
