```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Funcion(v,u,q):
    
    return u**q

def Solution(t,q):
    if q == 1:
        u = np.exp(t)
    else:
        u = ((1-q)*t+1)**(1/(1-q))        
    return u
```

```python
##metodo de Rugekutta sacado con las notas de clase de 1 ecuancion
def GetRungeKutta4(x,y,h,Funcion,q):
    f=Funcion
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1],q)
        k2 = f(x[i-1]+0.5*h,y[i-1]+0.5*h*k1,q)
        k3 = f(x[i-1]+0.5*h,y[i-1]+0.5*h*k2,q)
        k4 = f(x[i-1]+h,y[i-1]+h*k3,q)
        y[i]= y[i-1]+ h/6 *(k1+2*k2+2*k3+k4)
        
        
```

```python
N = 1000
t = np.linspace(0,10,N)
h = t[1] - t[0]
q = [0,0.2,0.4,0.7,0.9,1]

solucion = []

for i in range(len(q)):
    Y = np.zeros_like(t)
        
    if q[i] == 1:
        Y[0] = 1
        
    else:
        Y[0] = 1
        
    GetRungeKutta4(t,Y,h,Funcion,q[i])
    solucion.append(Y)
```

```python
Solex = []
Solex
for i in range(len(q)):
    Solex.append(Solution(t,q[i]))
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 5))

ax[0][0].plot(t,solucion[0],label="Solucion Numerica".format(q[0]))
ax[0][0].plot(t,Solex[0],label="Solucion Exacta".format(q[0]))
ax[0][0].set_title("q = 0")
ax[0][0].legend()

ax[0][1].plot(t,solucion[1],label="Solucion Numerica".format(q[1]))
ax[0][1].plot(t,Solex[1],label="Solucion Exacta".format(q[1]))
ax[0][1].set_title("q = 0.2")
ax[0][1].legend()

ax[0][2].plot(t,solucion[2],label="Solucion Numerica".format(q[2]))
ax[0][2].plot(t,Solex[2],label="Solucion Exacta".format(q[2]))
ax[0][2].set_title("q = 0.4")
ax[0][2].legend()

ax[1][0].plot(t,solucion[3],label="Solucion Numerica".format(q[3]))
ax[1][0].plot(t,Solex[3],label="Solucion Exacta".format(q[3]))
ax[1][0].set_title("q = 0.7")
ax[1][0].legend()

ax[1][1].plot(t,solucion[4],label="Solucion Numerica".format(q[4]))
ax[1][1].plot(t,Solex[4],label="Solucion Exacta".format(q[4]))
ax[1][1].set_title("q = 0.9")
ax[1][1].legend()

ax[1][2].plot(t,solucion[5],label="Solucion Numerica".format(q[5]))
ax[1][2].plot(t,Solex[5],label="Solucion Exacta".format(q[5]))
ax[1][2].set_title("q = 1")
ax[1][2].legend()

plt.tight_layout()
```
