```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
from tqdm import tqdm 
```

```python
def Funcion(x,omega):
    return -omega**2*x
```

```python
def MetodoVerlet(a,x0,v0,t,omega,h):   
    #siguiendo el codigo de clase
    x=np.zeros_like(t)
    v=np.zeros_like(t)
    x[0]=x0
    x[1]=x[0]+h*v0
    v[0]=v0
    v[1]=x[0]+h*a(x[1],omega)
    for i in range(2,len(x)):
        x[i]= 2*x[i-1]-x[i-2]+ a(x[i-1],omega)*h**2
        v[i] = (x[i]-x[i-2])/(2*h)
    return x,v
```

```python
omega=np.pi
h=0.1 
t=np.arange(0,20,h)
x0=1
v0=1
```

```python
x_esta,vx_esta= MetodoVerlet(Funcion,x0,v0,t,omega,h)
```

```python
scale=1
t1= t[::scale]
X_=x_esta[::scale]
VX=vx_esta[::scale]

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122)
#se inicializa el grÃ¡fico
def init():
    ax.clear()
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax1.clear()
    ax1.set_ylim(-2,2)
#se actualiza el grÃ¡fico       
def Update(i):   
    init()
    x = X_[i]
    vx =VX[i]
    circle = plt.Circle( (0,x), 0.8, fill=False )
    ax.add_patch(circle)   
    ax.arrow(0,x,0,vx,head_width=0.1,length_includes_head=True)
    ax1.plot( t1[0:i] , X_[0:i]) 
Animation = anim.FuncAnimation(fig,Update,frames=len(t1),init_func=init)
```
